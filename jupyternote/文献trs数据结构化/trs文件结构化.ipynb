```python
data = []
col_name = []
#python逐行读取txt文件
for line in open("CAE_Journal_Example32ab041e-5be3-4e80-a75d-0edce5c30252.trs","r",encoding='UTF-8'): #设置文件对象并读取每一行文件
    data.append(line.replace('=',':',1))  
line_num = 0
```

```python
import pandas as pd
```

```python
#读取列头为表头
for line_header in data:
    line_num=line_num+1
    print(line_num,line_header.split(':')[0])
    col_name_value = line_header.split(':')[0]
    col_name.append(col_name_value)
```

```python
colmn_header = list(set(col_name))
#print(colmn_header,type(colmn_header))
```

```python
#需要进一步优化为输入列名（按行进行输入）
df = pd.DataFrame(colmn_header)
print(df)
df.to_csv('test.csv')
print('done')
```
