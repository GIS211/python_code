```python
#!/usr/bin/python
# -*- coding: UTF-8 -*-
import os
import os.path

#
files_dir = r'E:\\'
file_list = os.listdir(files_dir)
print(file_list,type(file_list))


#txtutf-8
fobj=open(r'F:\resutl.txt','a',encoding='utf-8')

#
for file_name in file_list:
    
    #
    data = []
    
    #
    for line in open(files_dir + '\\' + file_name,"r",encoding='UTF-8'): #
        data.append(line.split('=',1))#data
    
    line_string = ''#
    #data
    for line_value in data:
        #
        if (len(line_value)==1): 
            fobj.write('\n' + line_string)
            line_string = ''
            continue
        #,<>
        line_string = line_string + '<>' + line_value[1].replace('\n','').replace('\r','')
#txt
fobj.close() 

print('done')
```
